# CS504 Project: Draft Success Analysis Data Collection

There were three different spreadsheets needed:
1) Team Draft History
2) Historical Team Records (Ws-Ls)
3) Individual Player Season Stats with AV Score

All of the data was collected from pro-football-reference.com.

Item 1, because of html oddities, was captured using the data scraping tool Octoparse.

Items 2 and 3 were captured using the Python Pandas library method read_html.

```python
import pandas as pd
```

## Collect Data

### Team Historical Records

```python
file = 'Draft_history_Football-links.xlsx'
filedf = pd.read_excel(file,sheet_name='team')
file_list = filedf.values.tolist()
endurl = '#all_team_index'
```

```python
#iterate over each team, collect their team success chart, and append it into one dataframe
i = 0
for team_url in file_list:
    beg_url= str(team_url).replace('[','').replace(']','').replace("'",'')
    
    if i == 0:
        
        df1 = pd.read_html(beg_url+endurl)
        df = df1[0]
        i += 1
    
    else:
        
        team_df = pd.read_html(beg_url+endurl)
        df = df.append(team_df[0]) 
        i += 1
```

```python
df
```

```python
df.to_excel('team_historical_records.xlsx')
```

### Player Season Data

```python
file2 = 'team_draft_data.xlsx'
filedf = pd.read_excel(file2,sheet_name='Sheet1')
url_list = filedf.player_link.values.tolist()
names = filedf.name.values.tolist()
names
```

```python
%%time

i = 0
for player_url in url_list:
    
    if len(str(player_url)) > 4:
    
        if i == 0:
        
            
            try:    
                df2 = pd.read_html(player_url)
                df3 = df2[1]
                df3['name'] = names[i]
                i += 1
            except:
                i += 1
                
    
        else:
        
            
            try:
                player_df = pd.read_html(player_url)
                player_df[1]['name'] = names[i]
                df3 = df3.append(player_df[1])
                i += 1
                    
            except:
                i += 1
                    
                
    else:
        
        i += 1
```

```python
df3
```

```python
df3.to_excel('player_data.xlsx')
```
